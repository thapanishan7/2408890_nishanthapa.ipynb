```python
#Subsetting By Indices: Examples
import pandas as pd
data = {"Name":["Alice","Bob","Charlie","David"],
        "Age": [24 , 19 , 22 , 25] ,
        "Score": [88 , 92 , 85 , 95]}
df = pd.DataFrame(data)

subset_iloc=df.iloc[1:3,0:2]
print(subset_iloc )

subset_loc=df.loc[df["Age"]>20,["Name","Score"]]
print(subset_loc )

subset_brackets = df[["Name","Age"]]
print(subset_brackets)
```

```python
#Subsetting by Values - Columns
import pandas as pd
data = {"Name":["Alice","Bob","Charlie","David"],
        "Age": [24 , 19 , 22 , 25] ,
        "Score": [88 , 92 , 85 , 95]}
df = pd.DataFrame(data)
name_column=df["Name"]
print(name_column)

name_and_age=df[["Name","Age"]]
print(name_and_age)
```

```python
#Subsetting by Rows (Filtering):
import pandas as pd

filtered_single=df[df["Age"]>20]
print(filtered_single)

filtered_multiple=df[(df["Age"]>20) & (df["Score"]>85)]
print(filtered_multiple)
```

```python
#Filtering on Categorical Values
import pandas as pd
data = {"Name":["Alice","Bob","Charlie","David"],
        "Department": ["HR" , "IT" , "Finance" , "HR"] ,
        "Score": [88 , 92 , 85 , 95]}
df = pd.DataFrame(data)
filtered_single=df[df["Department"]=="HR"]
print(filtered_single)

filtered_multiple=df[df["Department"].isin(["HR","IT"])]
print(filtered_multiple)
```

```python
#Group By Method - Apply - Aggregation:
import pandas as pd
data={"Category":["A","B","C","D","E"],
      "Value":[10,20,30,40,50]
      }
df=pd.DataFrame(data)
grouped=df.groupby("Category")["Value"].mean()
print(grouped)
```

```python
#Group By Method - Apply - Transformation:
import pandas as pd
data={"Category":["A","B","C","D","E"],
      "Value":[10,20,30,40,50]
      }
df=pd.DataFrame(data)
df["Normalized"]=df.groupby("Category")["Value"].transform(lambda x:x / x.sum())
print(df)
```

```python
#Group By Method - Apply - Filtration:
import pandas as pd
data={"Category":["A","B","A","B","A"],
      "Value":[10,20,30,40,50]
      }
df=pd.DataFrame(data)
filtered=df.groupby("Category").filter(lambda x:x ["Value"].sum()>60)
print(filtered)
```

```python
# 3.2 Exploratory Data Analysis Practice Exercise - 1.

# Which passenger had the highest fare paid relative to their age?
# To answer the question perform following operations:
# 1. Add a column to the Titanic dataset, fare per year, containing the fare divided by the age of
# the passenger(i.e., Fare/Age).
import pandas as pd
titanic=pd.read_csv("/content/drive/MyDrive/Python_dataset/Titanic-Dataset.csv")
titanic['Age'] = titanic['Age'].fillna(titanic['Age'].median())#kei missing xa vaye replace

# 1. Add a column to the Titanic dataset, fare per year, containing the fare divided by the age of
# the passenger(i.e., Fare/Age).
titanic['fare_per_year'] = titanic['Fare'] / titanic['Age']

# Subset rows where fare per year is higher than 5, assigning this to high fare age.
hage = titanic[titanic['fare_per_year'] > 5]

# Sort high fare age by descending fare per year, assigning this to high fare age srt.
hage_sort = hage.sort_values(by='fare_per_year', ascending=False)

# Select only the Name and fare per year columns of high fare age srt and save the result as result.
res = hage_sort[['Name', 'fare_per_year']]

print("Highest fare paid relative to their age:")
print(res)
```

```python
# Which adult male passenger (age ≥ 18 and Sex is ’male’) paid the highest fare relative to their
# class?
# To answer the question perform following operations:
# 1. Add a column to the Titanic dataset, fare per class, containing the fare divided by the passenger
# class i.e. Fare / Pclass.
import pandas as pd
titanic=pd.read_csv("/content/drive/MyDrive/Python_dataset/Titanic-Dataset.csv")
titanic['fare_per_class'] = titanic['Fare'] / titanic['Pclass']

# 2. Subset rows where the passenger is male (Sex is ”male”) and an adult (Age is greater than or equal
# to 18), assigning this to adult males.
adMale= titanic[(titanic['Sex'] == 'male') & (titanic['Age'] >= 18)]

# 3. Sort adult males by descending fare per class, assigning this to adult males srt.
adMale_sort = adMale.sort_values(by='fare_per_class', ascending=False)

# 4. Select only the Name, Age, and fare per class columns of adult males sr and save the result
# as result.
res1 = adMale_sort[['Name', 'Age', 'fare_per_class']]

# 5. Look at the result.
print("\nAdult male passenger (age ≥ 18 and Sex is ’male’) paid the highest fare relative to their class:")
print(res1)
```

```python
# 3.3 Exploratory Data Analysis with Group-by Method Practice Exercise:
# Based on the dataset Answer the following question:
# What percent of the total fare revenue came from each passenger class?
# To answer the question perform following operation:
# 1. Calculate the total Fare paid across all passengers in the Titanic dataset.
import pandas as pd
titanic=pd.read_csv("/content/drive/MyDrive/Python_dataset/Titanic-Dataset.csv")
titanic['Age'] = titanic['Age'].fillna(titanic['Age'].median())
total_fare = titanic['Fare'].sum()

# 2. Subset for passengers in first class (Pclass is 1) and calculate their total fare.
first_fare = titanic[titanic['Pclass'] == 1]['Fare'].sum()
second_fare = titanic[titanic['Pclass'] == 2]['Fare'].sum()
third_fare = titanic[titanic['Pclass'] == 3]['Fare'].sum()

# 3. Do the same for second class (Pclass is 2) and third class (Pclass is 3).
total = [first_fare, second_fare, third_fare]

# 4. Combine the fare totals from first, second, and third classes into a list.
combined = [fare / total_fare for fare in total]

# 5. Divide the totals for each class by the overall total fare to get the proportion of fare revenue by class.
print("Percentage of total fare revenue by class:")
class_labels = ['First Class', 'Second Class', 'Third Class']
for cls, prop in zip(class_labels, combined):
    print(f"{cls}: {prop * 100}%")
```

```python
# What percent of the total number of passengers on the Titanic belonged to each age group
# (e.g., child, adult, senior)?
# To answer the question perform following operation:
# 1. Create a new column, age group, that categorizes passengers into "child" (age < 18), "adult" (age
# 18{64), and "senior" (age 65 and above).
import pandas as pd
titanic=pd.read_csv("/content/drive/MyDrive/Python_dataset/Titanic-Dataset.csv")
titanic['fare_per_class'] = titanic['Fare'] / titanic['Pclass']
def categorize_age(age):
    if age < 18:
        return 'child'
    elif age < 65:
        return 'adult'
    else:
        return 'senior'
titanic['age_group'] = titanic['Age'].apply(categorize_age)

# 2. Calculate the total number of passengers on the Titanic.
total_passengers = len(titanic)

# 3. Count the number of passengers in each age group.
age_count = titanic['age_group'].value_counts()

# 4. Divide the count of each age group by the total number of passengers to get the proportion of passengers
# in each age group.
proportion = (age_count / total_passengers) * 100

# 5. Display the proportion as a percentage.
print("\nPercentage of passengers:")
for group, prop in proportion.items():
    print(f"{group.capitalize()}: {prop:.2f}%")
```
